Write a program to load the above given dataset, and print
top 5 and last 5 rows, and find the no. of row and no.
columns in the dataset.

```python
import pandas as pd
```

```python
df = pd.read_csv("./laptopPrice.csv")

df.reset_index(drop=True)
```

Find no. of entry for each “processor_brand”, and find which
processor brand laptop availability is high and low.


```python
counts = df['processor_brand'].value_counts()
print(counts)

highest_availability_brand = counts.idxmax()
lowest_availability_brand = counts.idxmin()

print(f"Highest availability: {highest_availability_brand}")
print(f"Lowest availability: {lowest_availability_brand}")
```

Make a group on “processor_brand” and find out the
minimum and maximum price of all the processor_brand
laptop.


```python

group = df.groupby('processor_brand')['Price'].agg(['min', 'max'])

print(group)
```

 Filter all the ‘APPLE’ brand laptop, and find top 5 APPLE
brand laptop which has the highest price. And top 5 those
has lowest price

```python
apple_laptop = df[df['brand'] == "APPLE"]

high_5 =apple_laptop.nlargest(5, 'Price')
low_5 =apple_laptop.nsmallest(5, 'Price')

print("Top 5 APPLE laptops with highest prices:")
print(high_5)

print("\nTop 5 APPLE laptops with lowest prices:")
print(low_5)


# In[10]:
```
