Write a program to find the minimum price of ktm bike,
which are available in Jaipur, and how many ktm bikes are
available at Jaipur location.

```python
import pandas as pd
```

```python
df2 =  pd.read_csv("./Used_Bikes.csv")
print(df2)
```

```python
df2.info
```

```python
filtered_df = df2[(df2['brand'] == 'KTM') & (df2['city'] == 'Jaipur')]
print(filtered_df)

min_price = filtered_df['price'].min()
print( "Minimum Price : ",min_price)

num = len(filtered_df)
print("Number of Bikes : ",num)
```

Find out how many bikes are available at Delhi location, and
that are “first owner” bikes. And also after filter the data
find out the minimum and maximum age of the bike.

```python
filter_data = df2[(df2['city'] == 'Delhi') & (df2['owner'] == 'First Owner') ]
print(filter_data)

max_price = filter_data['price'].max()
print( "Maximum Price : ",max_price)
min_price = filter_data['price'].min()
print("Minimum Price : ",min_price)
```
